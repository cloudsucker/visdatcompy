```python
import pandas as pd
from PIL import Image
from PIL.ExifTags import TAGS
from IPython.display import display
import os.path
```

```python
def get_exif_data(image_path):
    #exif_list = []
    exif_list = {}
    try:
        with Image.open(image_path) as img:
            exif_data = img._getexif()
            if exif_data:
                for tag, value in exif_data.items():
                    tag_name = TAGS.get(tag, tag)
                    exif_list[tag_name] = value
            else:
                print("No EXIF data found.")
    except FileNotFoundError:
        print("File not found.")

    return exif_list
```

```python
def getFilesFromDirectory(dir):
    filesList = []
    for address, dirs, files in os.walk(dir):
        for name in files:
            #проверка на формат файла
            if name.endswith('jpg'):
                filesList.append((name.split('.')[0], name.split('.')[-1], os.path.join(address, name)))
    return filesList
```

```python
def getExifTagsFromFiles(dir):
    filesExifList = []
    fileList = getFilesFromDirectory(dir)
    for file in fileList:
        filesExifList.append(({'Filename': file[0]} | {'FileExtension': file[1]} | get_exif_data(file[2])))
    return filesExifList
```

```python
dataPath = 'archive'
data = getExifTagsFromFiles(dataPath)
exiftags = list(TAGS.values())
df = pd.DataFrame(data = data,)
df = df.drop(columns='MakerNote')
df.to_csv('meta.csv', sep='\t', header=True, index=True)
display(df)
```
