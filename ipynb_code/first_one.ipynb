# Сканирование директории

```python
import os

def getFilesFromDirectory(dir):
    """
    ### Эта функция сканирует директорию и возвращает все файлы формата .jpg
        - Input: dir (path to file)
        - Output: Files List (1.jpg, 2.jpg, ...)
    """

    filesList = []
    for address, dirs, files in os.walk(dir):
        for name in files:
            # Проверка на формат файла
            if name.endswith('jpg'):
                filesList.append((name.split('.')[0], name.split('.')[-1], os.path.join(address, name)))
    return filesList
```

# Сбор и вывод метаданных файла

```python
import pandas as pd
from PIL import Image
from PIL.ExifTags import TAGS
from IPython.display import display

def get_exif_data(image_path):
    """
    ### Эта функция принимает на ввод путь к файлу формата '.jpg' и возвращает
    ### список метаданных изображения.
        - Input: image path
        - Output: list of exif data
    """

    exif_dict = {}
    banned_tags = ['MakerNote']

    try:
        with Image.open(image_path) as img:
            exif_data = img._getexif()
            if exif_data:
                for tag, value in exif_data.items():
                    tag_name = TAGS.get(tag, tag)
                    if (tag_name not in banned_tags):
                        exif_dict[tag_name] = value
            else:
                print("No EXIF data found.")
    except FileNotFoundError:
        print("File not found.")
    
    return exif_dict



# Укажите путь к вашему изображению
image_path = 'C:/Users/sharj/Desktop/Учёба/Учебная практика (5 сем.)/datasets/original_images/001.jpg'
exif_data = get_exif_data(image_path)

# Создание DataFrame из списка кортежей
df = pd.DataFrame.from_dict(exif_data, orient='index').T

display(df)

df.to_csv('meta.csv')
```
