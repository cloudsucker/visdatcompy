## **Сравнение методов поиска схожих изображений**

```python
import matplotlib.pyplot as plt

data = {
    'SIFT': 87.25,
    'AverageHash': 38.125,
    'PHash': 47.625,
    'MarrHildrethHash': 45.875,
    'RadialVarianceHash': 8,
    'BlockMeanHash': 53,
    'ColorMomentHash': 26.375,
}

sorted_data = dict(sorted(data.items(), key=lambda x: x[1], reverse=False))

# Определяем цвета для каждого метода
colors = []
for method in sorted_data.keys():
    if method == 'SIFT':
        colors.append('lightgreen')  # Цвет для метода дескрипторов
    else:
        colors.append('skyblue')  # Цвет для хэш-методов

fig, ax = plt.subplots()
bars = ax.barh(list(sorted_data.keys()), list(sorted_data.values()), color=colors)

ax.set_xlabel('Процент нахождения правильных пар')
ax.set_ylabel('Метод')

plt.title('Сравнение методов поиска схожих изображений')

# Добавление легенды
descriptors_patch = plt.Line2D([0], [0], color='lightgreen', lw=4, label='Дескрипторы')
hash_patch = plt.Line2D([0], [0], color='skyblue', lw=4, label='Хэш-функции')
plt.legend(handles=[descriptors_patch, hash_patch])

plt.show()
```
