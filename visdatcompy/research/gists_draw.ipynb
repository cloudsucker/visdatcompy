```python
import matplotlib.pyplot as plt

data = {
    'SIFT': 87.25,
    'AverageHash': 38.125,
    'PHash': 47.625,
    'MarrHildrethHash': 45.875,
    'RadialVarianceHash': 8,
    'BlockMeanHash': 53,
    'ColorMomentHash': 26.375,
}

sorted_data = dict(sorted(data.items(), key=lambda x: x[1], reverse=False))

fig, ax = plt.subplots()
ax.barh(list(sorted_data.keys()), list(sorted_data.values()), color='skyblue')

ax.set_xlabel('Процент нахождения правильных пар')
ax.set_ylabel('Метод')

plt.title('Сравнение методов поиска схожих изображений')

plt.show()
```
